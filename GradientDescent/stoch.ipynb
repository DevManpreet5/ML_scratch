```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
X,y = load_diabetes(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
lr=LinearRegression()
lr.fit(X_train,y_train)
print(lr.intercept_,lr.coef_)
```

```python
class SGDR:
    def __init__(self,epochs,lr):
        self.epochs=epochs
        self.lr=lr
    def fit(self,X_train,y_train):
        self.coeff=np.ones(X_train.shape[1])
        self.intercept=0

        for i in range(0,self.epochs):
            for _ in range(0,X_train.shape[0]):
                ind=np.random.randint(0,X_train.shape[0])
                y_hat=np.dot(X_train[ind],self.coeff)+self.intercept

                int_slope=-2*(y_train[ind]-y_hat)
                self.intercept=self.intercept-(self.lr*int_slope)

                coeff_slope=-2*np.dot((y_train[ind]-y_hat),X_train[ind])
                self.coeff=self.coeff-(self.lr*coeff_slope)
        print(self.intercept,self.coeff)

    def predict(self,X_test):
        return np.dot(X_test,self.coeff)+self.intercept

```

```python
sgdr=SGDR(epochs=10000,lr=0.001)
sgdr.fit(X_train,y_train)
```

```python
sgdr.predict(X_train)
```
